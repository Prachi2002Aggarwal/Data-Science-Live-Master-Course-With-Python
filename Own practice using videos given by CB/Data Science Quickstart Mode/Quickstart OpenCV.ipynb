```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('D:\Desktop\download.jpeg')
rgb_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
rgb_image
```

```python
rgb_image.shape
```

```python
# want to see how the image look like
plt.imshow(rgb_image)
# plt.axis("off")
plt.show()
```

```python
crop_image = rgb_image[50:200,50:150,:]
plt.imshow(crop_image)
plt.show()
```

```python
crop_image.shape
```

```python
total_pixels = crop_image.shape[0] * crop_image.shape[1]
```

```python
print(total_pixels)
```

```python
# resize your image in small
crop_image = cv2.resize(crop_image, (100,100))
```
