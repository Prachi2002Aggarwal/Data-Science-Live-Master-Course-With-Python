```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
subjects = "Maths" , "Chem" , "Physics" , "English"
weigtage = [20,10,15,5]
plt.pie(weigtage , labels = subjects)
plt.show()
```

```python
subjects = "Maths" , "Chem" , "Physics" , "English"
weigtage = [20,10,15,5]
plt.pie(weigtage , labels = subjects , explode = (0,0,1,0))
plt.show()
```

```python
subjects = "Maths" , "Chem" , "Physics" , "English"
weigtage = [20,10,15,5]
plt.pie(weigtage , labels = subjects , explode = (0,0,0.1,0))
plt.show()
```

```python
subjects = "Maths" , "Chem" , "Physics" , "English"
weigtage = [20,10,15,5]
plt.pie(weigtage , labels = subjects , explode = (0,0,0.1,0),autopct = '%1.1f%%')
plt.show()
```

```python
subjects = "Maths" , "Chem" , "Physics" , "English"
weigtage = [20,10,15,5]
plt.pie(weigtage , labels = subjects , explode = (0,0,0.1,0),autopct = '%1.1f%%', shadow = True)
plt.show()
```
