```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.bar([0,1,2],[10,20,15]) # current year
plt.bar([0,1,2],[20,10,12]) # next year
plt.show()
# now overlapping took place
```

```python
x = np.array([0,1,2]) * 2
plt.bar(x - 0.25,[10,20,15],width = 0.5 , label = "Current year", tick_label = ['a','b','c']) # current year
plt.bar(x+0.25 , [20,10,12] , width = 0.5 , label = "next yer=ar")
plt.legend()
plt.show()
```

```python
plt.style.use("dark_background")
x = np.array([0,1,2]) * 2
plt.bar(x - 0.25,[10,20,15],width = 0.5 , label = "Current year", tick_label = ['a','b','c'], color = 'pink') # current year
plt.bar(x+0.25 , [20,10,12] , width = 0.5 , label = "next year", color = 'yellow')
plt.ylim(0,30)
plt.legend()
plt.show()
```
