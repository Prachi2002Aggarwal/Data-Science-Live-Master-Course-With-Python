```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
themes = plt.style.available
print(themes)
```

```python
plt.style.use("seaborn")
```

```python
# Line Plot
x = np.arange(10)
y1 = x ** 2
y2 = 2 * x + 3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1 , color = 'red')
plt.show()
plt.plot(x,y2 , color = 'green')
plt.show()
```

```python
plt.plot(x,y1 , color = 'red' , label = "Apple",marker = 'o')
plt.plot(x,y2 , color = 'green', label = "Kiwi" , linestyle = "dashed",marker = "*")
plt.legend()
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fruits over Time")
plt.show()
```

```python
prices = np.array([1,2,3,4]) ** 2
print(prices)
```

```python
# 3 this automattically goes on y axes
# x apne aap 1,2,3,4 according to prices statement
plt.plot(prices)
plt.show()
```

```python
# Scatter Plot
plt.scatter(x,y1)
plt.scatter(x,y2)
plt.show()
```

```python
# Adjust the size of any plot
plt.figure(figsize=(2,2))
plt.scatter(x,y1 , color = 'red' , label = "Apple",marker = 'o')
plt.scatter(x,y2 , color = 'green', label = "Kiwi" , linestyle = "dashed",marker = "*")
plt.legend()
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fruits over Time")
plt.show()
```
