```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.tree import DecisionTreeClassifier , DecisionTreeRegressor
```

```python
X , y , centers = make_blobs (n_samples = 200 , n_features = 2 , centers = [[3 , 2] , [5 , 6]] , return_centers = True , random_state = 10)
print(X.shape , y.shape)
print(centers)
```

```python
plt.scatter(X[:,0] , X[:,1] , c = y)
plt.show()
```

```python
dt = DecisionTreeClassifier()
dt.fit(X,y)
```

```python
x1 = np.linspace(X[:,0].min() , X[:,0].max() , 50)
x2 = np.linspace(X[:,1].min() , X[:,1].max() , 50)

x1 , x2 = np.meshgrid(x1,x2)
x1 , x2 = x1.reshape(-1 , 1) , x2.reshape(-1 , 1)

x_test = np.concatenate([x1 , x2] , axis = 1)
print(x_test.shape)
```

```python
ypred = dt.predict(x_test)
```

```python
from matplotlib.colors import ListedColormap
colors = ["#bc84f5" , "#e8e49e"]
cmap = ListedColormap(colors)

plt.scatter(x_test[:,0] , x_test[:,1] , c = ypred , cmap = cmap)
plt.scatter(X[:,0] , X[:,1] , c = y)
plt.show()
```

```python
dt.score(X,y)
```
