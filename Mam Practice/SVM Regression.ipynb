```python
from sklearn.svm import LinearSVR, SVR
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, y = make_regression(n_samples=200,
    n_features=1, noise=20, bias=5, random_state = 30)
print(X.shape, y.shape)

plt.scatter(X, y)
plt.show()
```

```python
reg = LinearSVR()
reg.fit(X, y)

x_test = np.linspace(X.min(), X.max(), 50).reshape(-1,1)
ypred = reg.predict(x_test.reshape(-1,1))

plt.scatter(X, y)
plt.plot(x_test, ypred, c='crimson')
plt.show()
```

```python
# Making a new data
m = 50
x = np.linspace(-4, 4, m)
noise = np.random.randn(m,)
print(noise.shape)
y = x**2 + 1.8*noise + 3

plt.scatter(x, y)
plt.show()
```

```python
svr = SVR(kernel='poly', degree=2, C=1)
svr.fit(x.reshape(-1,1), y)
ypred = svr.predict(x.reshape(-1,1))
plt.scatter(x, y)
plt.plot(x, ypred, c='crimson')
plt.show()
```

```python
svr.score(x.reshape(-1,1), y)
```

```python
svr = SVR(kernel='poly', degree=2, C=10)
svr.fit(x.reshape(-1,1), y)
ypred = svr.predict(x.reshape(-1,1))
plt.scatter(x, y)
plt.plot(x, ypred, c='crimson')
plt.show()
```
