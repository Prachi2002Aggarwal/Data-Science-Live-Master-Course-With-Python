```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
X , y = make_moons(n_samples = 400 , noise = 0.11)
print(X.shape , y.shape)
```

```python
plt.scatter(X[:,0] , X[:,1])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(init = 'k-means++' , n_clusters = 4)
kmeans.fit(X)
ypred = kmeans.predict(X)
```

```python
plt.scatter(X[:,0] , X[:,1] , c = ypred)
plt.show()
```

```python
kmeans1 = KMeans(init = 'k-means++' , n_clusters = 2)
kmeans1.fit(X)
ypred1 = kmeans1.predict(X)
plt.scatter(X[:,0] , X[:,1] , c = ypred1)
plt.show()
```
