```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
X , y = make_moons(n_samples = 400 , noise = 0.11)
print(X.shape , y.shape)
```

```python
plt.scatter(X[:,0] , X[:,1])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(init = 'k-means++' , n_clusters = 4)
kmeans.fit(X)
ypred = kmeans.predict(X)
```

```python
plt.scatter(X[:,0] , X[:,1] , c = ypred)
plt.show()
```

```python
kmeans1 = KMeans(init = 'k-means++' , n_clusters = 2)
kmeans1.fit(X)
ypred1 = kmeans1.predict(X)
plt.scatter(X[:,0] , X[:,1] , c = ypred1)
plt.show()
```

```python
# We can say that K-Means is only goof for Linear data , bad for non linear data
# K - means : problem when we take less k than clusters , so sometimes cluster will not get center 
```

# DBSCAN

```python
# this do clustering according to density of points
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN (eps = 0.22 , min_samples = 4)
dbscan.fit(X)
```

```python
ypred2 = dbscan.labels_
```

```python
plt.scatter(X[:,0] , X[:,1] , c = ypred2)
plt.show()
```
