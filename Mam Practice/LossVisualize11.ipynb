```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from mpl_toolkits import mplot3d
```

```python
X , y , coef = make_regression(n_samples = 100,n_features = 1 , bias = 3 , noise = 5 , coef = True,random_state = 10)
```

```python
X.shape , y.shape
```

```python
y = y.reshape(-1,1)
```

```python
plt.scatter(X,y)
plt.show()
```

```python
def loss(X, y, M, C):
    yhat = (M*X.reshape(-1,)+C)
    residual_sq = (y.reshape(-1,) - yhat)**2
    J = residual_sq.sum(1)
    return J
```

```python
M = np.linspace(-100, 100, 200)
C = np.linspace(-50, 100, 200)
print(M.shape, C.shape)

M, C = np.meshgrid(M, C)
print(M.shape, C.shape)

M, C = M.reshape(-1,1), C.reshape(-1,1)
print(M.shape, C.shape)

J = loss(X, y, M, C)
print(J.shape)
```

```python
plt.figure(figsize=(7,7))
ax = plt.axes(projection='3d')
ax.scatter(M, C, J, c=J, cmap = plt.cm.coolwarm)
ax.set_xlabel("M (slope)")
ax.set_ylabel("C (bias)")
ax.set_zlabel("J (loss)")
plt.show()
```

```python

plt.figure(figsize=(7,7))
ax = plt.axes(projection='3d')
ax.scatter(M, C, J, c=J, cmap = plt.cm.coolwarm)
plt.show()
```

```python
residual = []
for m,c in zip(M,C):
    r = (y - (m*X+c)).reshape(-1,) #- > row vector of size (2,)
    residual.append(r)
print(len(residual), residual[0].shape)
```
