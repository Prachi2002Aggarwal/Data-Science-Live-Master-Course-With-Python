```python
# make a function for removing stop words
def remove_stopwords(words , sw) :
    useful_words =[]
    for word in words :
        if word.lower() not in sw:
            useful_words.append(word)
    return useful_words
```

```python
# now we call call the function
useful_words = remove_stopwords(words , stop_words)
print(len(useful_words))
print(useful_words)
```

```python
# 3) Lemmetization / Stemming
#      we have different words who have root word same so change everey word in root and remove unique
from nltk.stem import PorterStemmer , SnowballStemmer , LancasterStemmer
ps = PorterStemmer()
```

```python
# making a function for stemming
# we can only pass a single word so we need to make a function
def stemming(words , stemmer) :
    stemmed = [];
    for word in words :
        stemmed.append(stemmer.stem(word))
    return stemmed
```

```python
# call function
stemmed_words = stemming(useful_words , ps)
print(len(stemmed_words))
print(stemmed_words)
```

```python
# now as we can see we have root words but some words are repeated so we only want unique words
unique_words = set(stemmed_words)
print(len(unique_words))
# now stemmed_words is my vocabulary
```
