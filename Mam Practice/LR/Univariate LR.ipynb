```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "./data LR/"
X = np.load(path + "X_data.npy")
y = np.load(path + "Y_data.npy")
```

```python
X.shape , y.shape
```

```python
plt.scatter(X , y)
plt.xlabel("Feature")
plt.ylabel("Label/Output")
plt.title("Label vs Feature")
plt.show()
```

```python
m = ((X[:,0] * y).mean() - X[:,0].mean() * y.mean()) / ((X[:,0] ** 2).mean() - (X[:,0].mean()) ** 2)
```

```python
c = y.mean() - m * X[:,0].mean()
```

```python
x = np.linspace(X[:,0].min() , X[:,0].max() , 100)
print(x)
```

```python
y_hat = m * x + c
print(y_hat)
```

```python
plt.scatter(X,y,label = "Training Points")
plt.plot(x , y_hat , c = 'red' , label = "Best Fit Line")
plt.legend()
plt.xlabel("Feature")
plt.ylabel("Label/Output")
plt.show()
```
