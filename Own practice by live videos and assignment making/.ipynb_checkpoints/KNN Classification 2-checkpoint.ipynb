## KNN Class 

```python
# making kNN class
class kNN :
    def __init__(self , k = 3) :
        self.k = k
        
    def find_distance(self , pt1 , pt2) :
        return ((pt1[0] - pt2[0]) ** 2 + (pt1[1] - pt2[1]) ** 2) ** 0.5
    
    def predict(self , Xtrain , ytrain , Xtest) :
        distance = []
        for i in range(Xtrain.shape[0]) :
            dis = self.find_distance(Xtrain[i] , Xtest)
            distance.append(dis)
        k_index =  np.argsort(distance)[:self.k]
        y_req = y[k_index]
        y_req_unique , count = np.unique(y_req , return_counts = True)
        pred_class = y_req_unique[np.argmax(count)]
        return pred_class , count.max() / self.k  
```

```python
# taking one test point and checking our model
test_point = [1 , 5]
model = kNN()
prediction , accuracy = model.predict(X , y , test_point)
print(f"Predicted Class:{prediction}")
print(f"Accuracy :{accuracy}")
```
